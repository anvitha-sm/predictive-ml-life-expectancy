## Apply Winsorizations on Outliers

```python
for col in df.columns:
    wins_data = winsorize(df[col], limits=(0.05, 0.05))
    df.loc[:, col] = wins_data

for col in df.select_dtypes(include=np.number).columns:
    plt.figure()
    sns.boxplot(x=df[col]) # find outliers again
    plt.title(f"Boxplot of {col}")
    plt.show()
```

## Scaling All and Relevant Features

```python
y = df['Healthy_Life_Expectancy']
X_all = df.drop(columns=['Healthy_Life_Expectancy'])

scaler = StandardScaler()
numeric_cols = X_all.select_dtypes(include='number').columns
X_scaled_numeric = scaler.fit_transform(X_all[numeric_cols])
X_all = pd.DataFrame(X_scaled_numeric, columns=numeric_cols, index=X_all.index)

numeric_df = df.select_dtypes(include='number')
correlation = numeric_df.corr()['Healthy_Life_Expectancy'].abs()
irrelevant_features = correlation[correlation < 0.1].index.tolist()
relevant_features = correlation[correlation >= 0.1].index.tolist()
relevant_features.remove('Healthy_Life_Expectancy') 
X_reduced = df[relevant_features]
scaled_reduced = scaler.fit_transform(X_reduced)
X_relevant = pd.DataFrame(scaled_reduced, columns=relevant_features)
```

## Linear Regression on All and Relevant Features

```python
X_train_all, X_test_all, y_train, y_test = train_test_split(X_all, y, test_size=0.2, random_state=28)
X_train_rel, X_test_rel, _, _ = train_test_split(X_relevant, y, test_size=0.2, random_state=28)

lr = LinearRegression()
lr.fit(X_train_all, y_train)
y_pred_all = lr.predict(X_test_all)

print("Linear Regression with All Features")
print(f"MSE: {mean_squared_error(y_test, y_pred_all):.4f}")
print(f"R²: {r2_score(y_test, y_pred_all):.4f}")

sns.scatterplot(x=y_test, y=y_pred_all)
plt.plot(y_test, y_test, lw=2, label='Exact Predictions')
plt.xlabel("Test Y")
plt.ylabel("Prediction Y")
plt.title("Test vs Prediction Y")
plt.legend()

lr.fit(X_train_rel, y_train)
y_pred_rel = lr.predict(X_test_rel)
print("\nLinear Regression with Relevant Features (>0.1 corr)")
print(f"MSE: {mean_squared_error(y_test, y_pred_rel):.4f}")
print(f"R²: {r2_score(y_test, y_pred_rel):.4f}")
```

## GridSearch with Random Forest, XGBoost, and MLP with All Features

```python
rf_param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, 15],
    'min_samples_split': [2, 5]
}

xgb_param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.1, 0.05]
}

mlp_param_grid = {
    'hidden_layer_sizes': [(50,), (100,), (150,), (100, 100)],
    'activation': ['relu'],
    'solver': ['adam'],
    'max_iter': [2000]
}

rf_grid = GridSearchCV(RandomForestRegressor(random_state=28), rf_param_grid, cv=4, n_jobs=-1, scoring='neg_mean_squared_error')
xgb_grid = GridSearchCV(XGBRegressor(random_state=28), xgb_param_grid, cv=4, n_jobs=-1, scoring='neg_mean_squared_error')
mlp_grid = GridSearchCV(MLPRegressor(random_state=28), mlp_param_grid, cv=4, n_jobs=-1, scoring='neg_mean_squared_error')

rf_grid.fit(X_train_all, y_train)
xgb_grid.fit(X_train_all, y_train)
mlp_grid.fit(X_train_all, y_train)

best_rf = rf_grid.best_estimator_
best_xgb = xgb_grid.best_estimator_
best_mlp = mlp_grid.best_estimator_

print("Best Random Forest params:", rf_grid.best_params_)
print("Best XGBoost params:", xgb_grid.best_params_)
print("Best MLP params:", mlp_grid.best_params_)

def eval_model(name, model):
    y_pred = model.predict(X_test_all)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    print(f"{name} Test MSE: {mse:.4f}")
    print(f"{name} Test R²: {r2:.4f}\n")

print("\nTest Accuracy:")
eval_model("Random Forest", best_rf)
eval_model("XGBoost", best_xgb)
eval_model("MLP", best_mlp)
```
